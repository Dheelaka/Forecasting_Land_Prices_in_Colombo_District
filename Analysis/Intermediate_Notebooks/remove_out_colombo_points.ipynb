```python
import pandas as pd

# Load CSV files into Pandas DataFrames
df_a = pd.read_csv('../Dataset/New_Data/9_dist_cal.csv', encoding='latin1')
df_b = pd.read_csv('../Dataset/New_Data/within_colombo_only.csv', encoding='latin1')

A_new = df_a.copy()
A_new.loc[~A_new['ad_id'].isin(df_b['ad_id']), ['lat', 'lng']] = None

# Save the modified DataFrame A to a new CSV file
A_new.to_csv('../Dataset/New_Data/10_location_data_corrected.csv', index=False)
```
