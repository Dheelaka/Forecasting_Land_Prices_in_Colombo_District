```python
import pandas as pd

city_locations = pd.read_csv("../Dataset/New_Data/city_locations.csv", low_memory=False)
```

```python
import pandas as pd
from pyproj import Proj, transform

def convert_coordinates(lon, lat):
    in_proj = Proj(init='epsg:4326')  # Geographic coordinate system (WGS84)
    out_proj = Proj(init='epsg:5234')  # kandawala coordinate system

    # Convert coordinates
    x, y = transform(in_proj, out_proj, lon, lat)
    return x, y

city_locations[['x_kandawala', 'y_kandawala']] = city_locations.apply(lambda row: pd.Series(convert_coordinates(row['city_lon'], row['city_lat'])), axis=1)
```

```python
output_csv_path = '../Dataset/New_Data/city_locations.csv' 
city_locations.to_csv(output_csv_path, index=False)
```
