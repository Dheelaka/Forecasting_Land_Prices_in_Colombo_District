```python
import numpy as np
import pandas as pd
from tabulate import tabulate
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from fuzzywuzzy import fuzz
import re
```

```python
filtered_fro_IQ_Range_df = pd.read_csv("../Dataset/New_Data/5_zeros_removed_cleaned.csv", low_memory=False)
```

```python
df = filtered_fro_IQ_Range_df

def filter_iqr(group):
    q1 = group['price_land_pp'].quantile(0.25)
    q3 = group['price_land_pp'].quantile(0.75)
    return group[(group['price_land_pp'] >= q1) & (group['price_land_pp'] <= q3)]

filtered_IQRange_df_1 = df.groupby('main_city').apply(filter_iqr).reset_index(drop=True)
```

```python
df = filtered_IQRange_df_1

def filter_iqr(group):
    q1 = group['price_land_pp'].quantile(0.25)
    q3 = group['price_land_pp'].quantile(0.75)
    return group[(group['price_land_pp'] >= q1) & (group['price_land_pp'] <= q3)]

filtered_IQRange_df_2 = df.groupby('main_city').apply(filter_iqr).reset_index(drop=True)
```

```python
df = filtered_IQRange_df_2

def filter_iqr(group):
    q1 = group['price_land_pp'].quantile(0.25)
    q3 = group['price_land_pp'].quantile(0.75)
    return group[(group['price_land_pp'] >= q1) & (group['price_land_pp'] <= q3)]

filtered_IQRange_df_3 = df.groupby('main_city').apply(filter_iqr).reset_index(drop=True)
```

```python
filtered_IQRange_df_3
```

```python
aggregations = {
    'price_land_pp': ['min', 'max', 'mean']
}

price_by_city = filtered_IQRange_df_3.groupby('main_city').agg(aggregations)

price_by_city.columns = ['min_price', 'max_price', 'average_price']

price_by_city.reset_index(inplace=True)

price_by_city.fillna(0, inplace=True)  

pd.set_option('display.max_rows', None)

def format_millions(x):
    return '{:.3f}Mn'.format(x / 1e6)

pd.set_option('display.float_format', format_millions)

price_by_city
```
