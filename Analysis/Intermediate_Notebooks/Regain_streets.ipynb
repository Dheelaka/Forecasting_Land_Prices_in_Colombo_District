```python
import pandas as pd

# Load CSV files into dataframes
df_a = pd.read_csv("../Dataset/New_Data/7_price_filtered.csv")
df_b = pd.read_csv('../Dataset/New_Data/1_2018-2023_combined.csv')

# Merge dataframes on 'ad_id'
merged_df = pd.merge(df_a, df_b, on='ad_id', how='left')

# Create a new column 'streets' in df_a and fill it with 'street' values from df_b
df_a['streets'] = merged_df['street']

# Find the index of the 'city' column
city_index = df_a.columns.get_loc('city')

# Reorder columns to place 'streets' before 'city'
df_a = pd.concat([df_a.iloc[:, :city_index], df_a[['streets']], df_a.iloc[:, city_index:]], axis=1)

output_csv_path = '../Dataset/New_Data/with_streets.csv' 
df_a.to_csv(output_csv_path, index=False)

# # Save the updated dataframe with the new column
# df_a.to_csv('updated_file_a.csv', index=False)
```
