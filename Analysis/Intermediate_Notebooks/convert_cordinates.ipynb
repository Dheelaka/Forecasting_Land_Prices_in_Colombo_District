# Convert Cordinates

```python
import pandas as pd
from pyproj import Proj, transform

def convert_coordinates(lon, lat):
    in_proj = Proj(init='epsg:4326')  # Geographic coordinate system (WGS84)
    out_proj = Proj(init='epsg:5234')  # SLD99 projected coordinate system

    # Convert coordinates
    x, y = transform(in_proj, out_proj, lon, lat)
    return x, y

input_csv_path = '../Dataset/New_Data/city_locations.csv'
df = pd.read_csv(input_csv_path)

df[['x_kandawala', 'y_kandawala']] = df.apply(lambda row: pd.Series(convert_coordinates(row['city_lon'], row['city_lat'])), axis=1)

output_csv_path = '../Dataset/New_Data/7.2_city_average_prices_with_cordinates.csv' 
df.to_csv(output_csv_path, index=False)

print(f"Conversion completed. Result saved to {output_csv_path}")
```
