## Calculate Price Per Perch

```python
def calculate_land_price(row):
    if row['price_land_total'] == 0 or pd.isnull(row['price_land_total']) or row['land_perch'] == 0 or pd.isnull(row['land_perch']):
        return np.nan
    else:
        return row['price_land_total'] / row['land_perch']

# Apply the function to create the land_price_PP column
bare_land_df_na_dropped_cleaned['Gen_land_price_PP'] = bare_land_df_na_dropped_cleaned.apply(calculate_land_price, axis=1) #Gen means generated column
```
